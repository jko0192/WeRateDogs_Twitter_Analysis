# Insights And Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

df = pd.read_csv(r'twitter_archive_master.csv')
```

## Univariate Exploration

```python
variables = ['rating_numerator', 'rating_denominator', 'favorite_count', 'retweet_count']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=[15,10])
axes = axes.flatten()

for index, variable in enumerate(variables):
    min_value = 0
    max_value = np.log10(df[variable].max())
    bin_size = (max_value - min_value) / 50
    bins = 10 ** np.arange(min_value, max_value + bin_size, bin_size)
    xlabel = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Count', '').strip()
    title = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Count', '').strip()
    color = sb.color_palette(palette='pastel')[0]
    
    if variable == 'rating_numerator' or variable == 'rating_denominator':
        tick_locations = [1e0, 1e1, 1e2, 1e3]
        tick_labels = [1, 10, 100, 1000]
        axes[index].hist(data=df, x=variable, bins=bins, color=color)
        axes[index].set_xscale('log')
        axes[index].set_xticks(tick_locations)
        axes[index].set_xticklabels(tick_labels, fontsize=12)
        axes[index].set_xlabel(xlabel, fontsize=12)
    else:
        tick_locations = [1e0, 1e1, 1e2, 1e3, 1e4, 1e5]
        tick_labels = [1, 10, 100, 1000, 10000, 100000]
        axes[index].hist(data=df, x=variable, bins=bins, color=color)
        axes[index].set_xscale('log')
        axes[index].set_xticks(tick_locations)
        axes[index].set_xticklabels(tick_labels, fontsize=12)
        axes[index].set_xlabel(xlabel, fontsize=12)
    
    axes[index].set_ylabel('Frequency', fontsize=12)
    axes[index].set_title('{} By Frequency'.format(title), fontsize=12);
    
fig.tight_layout(pad=3);
```

### Frequency Distributions For Ratings, Favorites, and Retweets
> The frequency distribution for numerator ratings appears unimodal and slightly left-skewed after a log transformation on the x-axis, while the frequency distribution for denominator ratings is discrete

> The frequency distributions for favorites and retweets follow a log normal distribution after a log transformation on the x-axis. The frequency distribution for retweets appears slightly right-skewed with the bulk of the distribution between 1,000 and 10,000 retweets. On the other hand, the bulk of the distribution for favorites is between about 8,000 to 50,000 favorites

```python
variables = ['sources', 'stage']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[15,8])
axes = axes.flatten()

for index, variable in enumerate(variables):
    color = sb.color_palette(palette='pastel')[1]
    if variable == 'stage': 
        query = df[df[variable] != '']
        order = query[variable].value_counts().index
        sb.countplot(data=query, y=variable, ax=axes[index], color=color, order=order)
    else:
        order = df[variable].value_counts().index
        sb.countplot(data=df, y=variable, ax=axes[index], color=color, order=order)
    
    axes[index].set_xlabel('Frequency', fontsize=12)
    axes[index].set_ylabel(variable.capitalize(), fontsize=12)
    axes[index].set_title('{} By Frequency'.format(variable.capitalize()));
        
fig.tight_layout(pad=5.0)
```

### Frequency Distributions For Sources And Stages
> The frequency distribution for sources shows twitter for iphone with the highest frequency and tweet deck with the lowest frequency. This indicates the bulk of distribution for tweets are from iphones as its most frequent source

> The frequency distribution for stages shows pupper with the highest frequency and doggo with the second to highest frequency. This indicates the bulk of distribution for tweets included pupper more frequently than those of other stages

## Bivariate Exploration

```python
variables = ['rating_numerator', 'rating_denominator', 'favorite_count', 'retweet_count']
plt.figure(figsize=[10,8])

heatmap = sb.heatmap(df[variables].corr(), annot=True, fmt='.3f', vmin=0, cmap='mako')
heatmap.set_title('Pairwise Correlation Between Numeric Attributes');
```

```python
fig = plt.figure(figsize=[10,8])
color = sb.color_palette(palette='pastel')[3]
plt.scatter(data=df, x='retweet_count', y='favorite_count', color=color, alpha=0.7)
plt.xlim(0,8000)
plt.ylim(0,30000)
plt.xlabel('Retweet', fontsize=12)
plt.ylabel('Favorite', fontsize=12)
plt.title('Pairwise Correlation Between Retweet And Favorite', fontsize=15);
```

### Pairwise Correlation Between Numeric Attributes
> The pairwise correlation shows a strong positive relationship between favorites and retweets with correlation values of 0.94. This strongly suggests that as one variable increases, the other variable also increases. On the other hand, numerator ratings and denominator ratings show weak relationships with correlation values of less than 0.2.

> The scatter plot correlation between retweets and favorites confirms their strong positive relationship indicating as the number of retweets increases, the number of favorites also increases

```python
variables = [('sources','favorite_count'),('sources','retweet_count'),('stage', 'favorite_count'),('stage', 'retweet_count')]
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=[20,15])
axes = axes.flatten()

for index, variable in enumerate(variables):
    color = sb.color_palette(palette='pastel')[4]
    if variable[0] == 'stage':
        xlabel = variable[0].capitalize()
        ylabel = ' '.join([string.capitalize() for string in variable[1].split('_')]).replace('Count', '').strip()
        query = df[df[variable[0]] != '']
        
        sb.boxplot(data=query, x=variable[0], y=variable[1], ax=axes[index], color=color)
        axes[index].tick_params(axis='x', labelsize=13, rotation=60)
        axes[index].tick_params(axis='y', labelsize=13)
        axes[index].set_xlabel(xlabel, fontsize=15)
        axes[index].set_ylabel(ylabel, fontsize=15)
        axes[index].set_title('Pairwise Correlation Between {} And {}'.format(ylabel, xlabel), fontsize=15)
    else:
        xlabel = variable[0].capitalize()
        ylabel = ' '.join([string.capitalize() for string in variable[1].split('_')]).replace('Count', '').strip()
        
        sb.boxplot(data=df, x=variable[0], y=variable[1], ax=axes[index], color=color)
        axes[index].tick_params(axis='x', labelsize=13, rotation=60)
        axes[index].tick_params(axis='y', labelsize=13)
        axes[index].set_xlabel(xlabel, fontsize=15)
        axes[index].set_ylabel(ylabel, fontsize=15)
        axes[index].set_title('Pairwise Correlation Between {} And {}'.format(ylabel, xlabel), fontsize=15);
        
fig.tight_layout(pad=5)
```

### Pairwise Correlation With Sources And Stages
> The pairwise correlation with sources shows twitter for iphone captures the bulk distribution of favorites and the bulk distribution of retweets. This indicates twitter for iphone captures more favorites and retweets than those of other sources. Tweet deck captures the least favorites and retweets amongst all sources

> The pairwise correlation with stages shows tweets with puppo have more favorites and retweets than those of other stages. Puppo also has a larger interquartile range than those of other stages 

```python
variables = ['retweet_count', 'favorite_count']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[20,10])
axes = axes.flatten()

for index, variable in enumerate(variables):
    color = sb.color_palette(palette='pastel')[5]
    xlabel = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Count', '(% proportion)')
    title = ' '.join([string.capitalize() for string in variable.split('_')]).replace('Count', '').strip()
    query = df.groupby(['month'])[variable].sum()
    total = df[variable].sum()
    prop = (query / total).reset_index().sort_values(by=variable, ascending=False)
    
    barplot = sb.barplot(data=prop, x=variable, y='month', estimator=sum, ax=axes[index], color=color)
    barplot.set_xlabel(xlabel, fontsize=15)
    barplot.set_ylabel('Month', fontsize=15)
    barplot.tick_params(labelsize=15)
    barplot.set_title('Month By {} Proportion'.format(title), fontsize=15)

fig.tight_layout(pad=3)
```

### Months By Retweets And Favorites Proportion
> The month of January has the most retweets in proportion to total retweets across all months, while August has the least retweets in proportion. In addition, the month of July has the most favorites in proportion to total favorites across all months, while August has the least favorites in proportion
